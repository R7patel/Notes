<a name="MissindDataImputation"></a><h1>Missind Data Imputation</h1>

<ol>
    <li><strong>Complete Case Analysis:</strong>
        <ul>
            <li>Data is missing completely at randome</li>
            <li>Delete all records with missing data</li>
            <li>No more than 5% of total dataset contains missing data</li>
        </ul>
    </li>
    <li><strong>Mean and Median imputation:</strong>
        <ul>
            <li>Data is missing completely at randome</li>
            <li>The missing data mostlikely looks like majority of the observations</li>
            <li>Hire percentage of NA results in higher distortation of distribution and varience (IQRÂ increases due to imputation =leads to=>Â an apparent increase in outliers as Vickers narrows)</li>
            <li>No more than 5% of variable contains missing data</li>
            <li>Sometimes used along with missing_indicator_flag</li>
        </ul>
    </li>
    <li><strong>Mode imputation:</strong>
        <ul>
            <li>Data is missing completely at randome</li>                  
            <li>No more than 5% of variable contains missing data</li>
            <li>The missing data most likely looks like majority of the observations</li>
            <li>Hire percentage of NA results in higher distortation</li>
            <li>Over representation of majorioty class/value</li>
        </ul>  
    </li>
    <li><strong>Arbitrary Value imputation:</strong>
        <ul>
            <li><strong>Numerical:</strong>0,-1,999,-999; <strong>Categorical:</strong>"MISSING","UNKNOWN","NA",..</li>
            <li>Edge values of distributions, not in middle</li>
            <li>Data is not missing at randome</li>
            <li>Hire percentage of NA results in higher distortation</li>
                <li>Makes it look like there are less outliers</li>
            <li>Ok for tree Algo but not ok for linear models</li>
            <li>End of distribution imputation (Only for numerical variable, impute using value at the end of distribution)</li>
                <ul> 
                    <li>Mean +- ( {3 or 1.5} x SD)</li>   
                    <li>Masks outliers, affects varience</li>   
                </ul>
        </ul>
    </li>
    <li><strong>Randome Sample imputation:</strong>
        <ul>
            <li>Filling the missing data with values selected completely at randome from the available observed values</li>        
        </ul>  
    </li>
</ol>

<a name="OutlierHandling"></a><h1>Outlier Handling</h1>

<h3>Effects of Outliers in data</h3>
- All linear models and AdaBoost techinques are affected by outliers in the data

<h3>Detecting Outliers in data</h3>

<ol>
    <li><strong>Gaussian distribution (mean and std):</strong>
        <ul>
            <li><strong>Normal distribution:</strong> For Normal distribution <strong>~99%</strong> of observations are within <strong>mean +- 3 x Standard Deviations</strong>, so values outside are considered outliers</li>
        </ul>
    </li>
    <li><strong>Inter-quantal range proximity rule:</strong></li>
        <ul>
            <li>For <strong>NON Normal distribution</strong>  we need to calculate Inter-quantile range (IQR) and the quantiles,as follow: </li>
            <li>IQR = 75th Quantile - 25th Quantile</li>
            <li><strong>Upper Limit = 75th Quantile + IQR x 1.5</strong></li>
            <li><strong>Lower Limit = 25th Quantile - IQR x 1.5</strong></li>
            <li>So the values outside of upper and lower limit are considered outliers.</li>
            <li><strong>NOTE:</strong> for extreme outliers, multiply the IQR by 3 instead of 1.5</li>
        </ul>
    <li><strong>Quantiles:</strong></li>
        <ul>
            <li>For <strong>Normal distribution</strong> we need to calculate the 5th quantile and 95th quantile</li>
            <li>There is ~95% of data above 5th quantile and there is ~95% of data below 95th quantile.</li>
            <li>So the values outside of <strong>5th quantile and 95th quantile</strong> are considered outliers.</li>            
        </ul>
</ol>

<h3>Handling Outliers Techniques</h3>

<ol>
    <li><strong>Trimming:</strong>
        <ul>
            <li>Detect & remove outliers from data (Delete rows).</li>
            <li> Outliers are rare so theoritically very less % of data is deleted.</li>
            <li><strong>Pro: </strong> simple and fase</li>
            <li><strong>Cons: </strong> Big chunk of data is removed if outliers are more.</li>
        </ul>
    </li>
    <li><strong>Missing Data:</strong>
        <ul>
            <li>Assume the outlier value is missing (set as NAN)</li>
            <li>Use missing data imputaion on those values</li>
        </ul>
    </li>
        <li><strong>Descritization:</strong>
        <ul>
            <li>Put outliers in lower or upper bins </li>
            <li>Just like descritization/binning but a column have only one bin</li>
        </ul>
    </li>
    <li><strong>Censoring:</strong>
        <ul>
            <li>Capping
                <ul>
                    <li>Cap minimun and maximum for a perticular variable</li>
                    <li>So replace outlier(extreme values) with border values</li>
                    <li><strong>Pro: </strong> No data is removed</li>
                    <li><strong>Cons: </strong> Variable distribution is distorted</li>
                </ul></li>
            <li>Top / Bottom coding </li>
            <li>Winsorization (SkLearn Feature-engine) - [Pending]</li>
        </ul>
    </li>
</ol>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


def create_ages(mu=50,sigma=13, num_samples=1000, seed=42):
    np.random.seed(seed)
    sample_ages=np.random.normal(loc=mu, scale=sigma,size=num_samples)
    sample_ages=np.round(sample_ages,decimals=0)
    return sample_ages

age_data=create_ages()
df = pd.DataFrame(data= age_data,columns=['Age'])
df['Age'] =df['Age'].astype(int)
df
```
